```python
names_2016 = ['Company_size',
                    'Is_tech',
                    'MentalH_insurance',
                    'MentalH_company_info',
                    'MentalH_company_resources',
                    'Anonymity_protection',
                    'Medical_leave_MH',
                    'Superviser_comfortable_talking',
                    'Coworkers_comfortable_talking',
                    'Have_MHD',
                    'Past_MHD',
                    'Treatment_sought',
                    'Work_interference_with_Treatment',
                    'Work_interference_without_Treatment',
                    'Discouranged_in_revealing_MHD_at_work',
              'MHD_identification_effect_on_career',
                    'Age',
                    'Gender']
```

```python
data_2016 = data_2016.rename(columns=dict(zip(keep_2016,names_2016)))
data_2016.head()
```

```python
drop_2016 = [i for i in data_2016.columns if not (i in names_2016)]
data_2016 = data_2016.drop(columns = drop_2016)
data_2016.head()
```

```python
data_2016 = data_2016.replace({'Company_size':{'1-5':'1','6-25':'2','26-100':'3','100-500':'4','500-1000':'5','More than 1000':'6'}})
data_2016 = data_2016.replace({'MentalH_insurance':{'Yes':'1','I don\'t know':'2','No':'0','Not eligible for coverage / NA':'3'}})
data_2016 = data_2016.replace({'MentalH_company_info':{'Yes':'1','I don\'t know':'2','No':'0'}})
data_2016 = data_2016.replace({'MentalH_company_resources':{'Yes':'1','I don\'t know':'2','No':'0'}})
data_2016 = data_2016.replace({'Anonymity_protection':{'Yes':'1','I don\'t know':'2','No':'0'}})
data_2016 = data_2016.replace({'Medical_leave_MH':{'Difficult':'0','Somewhat difficult':'1','Neither easy nor difficult':'2','Somewhat easy':'3','Very easy':'4','I don\'t know':'-1'}})
data_2016 = data_2016.replace({'Superviser_comfortable_talking':{'Yes':'2','Maybe':'1','No':'0'}})
data_2016 = data_2016.replace({'Coworkers_comfortable_talking':{'Yes':'2','Maybe':'1','No':'0'}})
data_2016 = data_2016.replace({'Have_MHD':{'Yes':'1','Maybe':'1','Don\'t Know':'2','No':'0'}})
data_2016 = data_2016.replace({'Past_MHD':{'Yes':'1','Maybe':'1','Don\'t Know':'2','No':'0'}})
data_2016 = data_2016.replace({'Family_history_MHD':{'Yes':'1','I don\'t know':'2','No':'0'}})
data_2016 = data_2016.replace({'Work_interference_with_Treatment':{'Often':'1','Sometimes':'2','Rarely':'3','Never':'4','Not applicable to me':'-1'}})
data_2016 = data_2016.replace({'Work_interference_without_Treatment':{'Often':'1','Sometimes':'2','Rarely':'3','Never':'4','Not applicable to me':'-1'}})
data_2016 = data_2016.replace({'Discouranged_in_revealing_MHD_at_work':{'Yes':'2','Maybe':'1','No':'0'}})
data_2016 = data_2016.replace({'Gender':{'Male':'0','male':'0','m':'0','Let\'s keep it simple and say "male"':'0','M':'0','Male ':'0','Man':'0','Make':'0',
                                'Female':'1','female':'1','F':'1','Woman':'1','f':'1','Female ':'1',
                                'Cishet male':'2','Nonbinary':'2','agender':'2','Questioning':'2','Cis Male':'2','cis woman':'2','Agender trans woman':'2','Trans man':'2','None':'2','Trans non-binary/genderfluid':'2','CIS Male':'2','Female (cis)':'2'}})
data_2016 = data_2016.replace({'MHD_identification_effect_on_career':{'Yes, I think they would':'10','Maybe':'5',"No, I don't think they would":'0'}})
data_2016.head()
```

```python
def fixNan(colList,df,value):
    for icol in colList:
        if (len(df[icol].isna()) > 0):
            df[icol].fillna(value, inplace=True)
            #df = df.replace({colName:{'nan':value}})
    return df

nan_to_min1 = ['Company_size', 'Medical_leave_MH',
                       'Have_MHD', 'Past_MHD', 'Treatment_sought', 'Discouranged_in_revealing_MHD_at_work',
                       'Age', 'Gender', 'Work_interference_with_Treatment', 'Work_interference_without_Treatment']
nan_to_1 = ['Superviser_comfortable_talking', 'Coworkers_comfortable_talking', 'MHD_identification_effect_on_career']
nan_to_2 = ['MentalH_insurance', 'MentalH_company_info', 'MentalH_company_resources','Anonymity_protection']
drop_nan = []

data_2016 = fixNan(nan_to_min1,data_2016,'-1')
data_2016 = fixNan(nan_to_1,data_2016,'1')
data_2016 = fixNan(nan_to_2,data_2016,'2')

for icol in drop_nan:
    data_2016 = data_2016[data_2016[icol].notna()]
    
data_2016.head()    
```
